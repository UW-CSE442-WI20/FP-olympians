```python
import pandas as pd
import numpy as np
import re

regex = re.compile('[^a-zA-Z ]')
```

```python
olympics = pd.read_csv('athlete_events.csv')
```

```python
olympics.head()
```

```python
olympics.shape
```

```python
recent = olympics[olympics['Year'] >= 1980]

recent.shape
```

```python
final = recent[recent['Season'] == 'Summer']

final.shape
```

```python
final
```

```python
final['Team'] = final['Team'].apply(lambda x: re.sub(regex, '', str(x)))
```
